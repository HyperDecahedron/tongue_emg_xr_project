```python
# -------------------------- Visualization of Raw data
import pandas as pd
import matplotlib.pyplot as plt

# Load OpenBCI data
path = "C:/Quick_Disk/tonge_emg_xr_project/data/Recordings/"
file_name = "recording1.csv"
df = pd.read_csv(path + file_name, sep='\t', skiprows=5, engine='python')

# Define sampling rate and duration
sampling_rate = 250  # Hz
duration_sec = 20
num_samples = sampling_rate * duration_sec  # 5000 samples

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(12, 10), sharex=True)

# Plot Channel 1
axs[0].plot(df.iloc[:num_samples, 1], color='b')
axs[0].set_title("Channel 1 - First Seconds")
axs[0].set_ylabel("Voltage (mV)")
axs[0].grid(True)

# Plot Channel 2
axs[1].plot(df.iloc[:num_samples, 2], color='g')
axs[1].set_title("Channel 2 - First Seconds")
axs[1].set_ylabel("Voltage (mV)")
axs[1].grid(True)

# Plot Channel 3
axs[2].plot(df.iloc[:num_samples, 3], color='r')
axs[2].set_title("Channel 3 - First Seconds")
axs[2].set_xlabel("Sample Index (at 250 Hz)")
axs[2].set_ylabel("Voltage (mV)")
axs[2].grid(True)

plt.tight_layout()
plt.show()
```
