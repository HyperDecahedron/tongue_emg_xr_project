```python
# Load data with no header row, force numeric column names
df = pd.read_csv("Test1Swallow.csv", sep='\t', skiprows=5, header=None)

# Rename for clarity
df.columns = [f"ch_{i}" for i in range(df.shape[1])]

# Print new column names
print(df.columns.tolist())
```

```python
plt.figure(figsize=(12, 4))
plt.plot(df["ch_1"][7500:8500])  # zoom around 00:31 (swallow 1)
plt.title("Zoomed-In View of EMG (Swallow 1, Samples 75008500)")
plt.xlabel("Sample Index")
plt.ylabel("Voltage")
plt.grid(True)
plt.show()
```

```python
from scipy.signal import butter, filtfilt

def bandpass_filter(data, lowcut=20.0, highcut=100.0, fs=250.0, order=4):
    nyq = 0.5 * fs  # 125 Hz
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return filtfilt(b, a, data)

# 2. Apply to your EMG data (channel 1)
emg_raw = df["ch_1"]
emg_filtered = bandpass_filter(emg_raw)

# 3. Plot a zoomed-in section before & after
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(emg_raw[7500:8500], label="Raw", color='gray')
plt.title("Raw EMG")
plt.grid(True)

plt.subplot(2, 1, 2)
plt.plot(emg_filtered[7500:8500], label="Filtered", color='blue')
plt.title("Filtered EMG (20450 Hz Bandpass)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
swallow_indices = [7750, 12250, 17250, 20000, 25000, 29000, 31500, 36250, 42500]
window_size = 250  # 1 second before and after
swallow_windows = []

for idx in swallow_indices:
    start = idx - window_size
    end = idx + window_size
    if start >= 0 and end < len(emg_filtered):
        window = emg_filtered[start:end]
        window = pd.Series(window).reset_index(drop=True)
        swallow_windows.append(window)
```

```python
print(swallow_windows[0].shape)  # should be (500,)
```

```python
import numpy as np

# Root Mean Square
def rms(signal):
    return np.sqrt(np.mean(signal**2))

# Zero Crossings
def zero_crossings(signal):
    return np.sum(np.diff(np.sign(signal)) != 0)

# Waveform Length
def waveform_length(signal):
    return np.sum(np.abs(np.diff(signal)))
```

```python
features = []

for window in swallow_windows:
    f_rms = rms(window)
    f_zc = zero_crossings(window)
    f_wl = waveform_length(window)
    features.append([f_rms, f_zc, f_wl])
```

```python
import pandas as pd

X = pd.DataFrame(features, columns=["RMS", "ZC", "WL"])
y = [1] * len(X)  # all swallows are class 1 for now

print(X)
```

```python
rest_indices = [2000, 4000, 10000, 15000, 22000, 27000, 32000, 37000, 46000]  # pick these manually or with care
rest_windows = []

for idx in rest_indices:
    start = idx - window_size
    end = idx + window_size
    if start >= 0 and end < len(emg_filtered):
        window = pd.Series(emg_filtered[start:end]).reset_index(drop=True)
        rest_windows.append(window)
```

```python
rest_features = []

for window in rest_windows:
    f_rms = rms(window)
    f_zc = zero_crossings(window)
    f_wl = waveform_length(window)
    rest_features.append([f_rms, f_zc, f_wl])
```

```python
X_rest = pd.DataFrame(rest_features, columns=["RMS", "ZC", "WL"])
y_rest = [0] * len(X_rest)

# Combine with swallow features
X_full = pd.concat([X, X_rest], ignore_index=True)
y_full = y + y_rest
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns

# 1. Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_full, y_full, test_size=0.3, random_state=42
)

# 2. Train model
clf = RandomForestClassifier(n_estimators=100, random_state=42)
clf.fit(X_train, y_train)

# 3. Predict
y_pred = clf.predict(X_test)

# 4. Evaluate
acc = accuracy_score(y_test, y_pred)
cm = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {acc:.2f}")

# 5. Plot confusion matrix
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=["Rest", "Swallow"], yticklabels=["Rest", "Swallow"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()
```

```python
import matplotlib.pyplot as plt

importances = clf.feature_importances_
feature_names = X.columns

plt.bar(feature_names, importances)
plt.title("Feature Importance")
plt.ylabel("Importance Score")
plt.grid(True)
plt.show()
```

```python
# THIS WILL NOT WORK OF COURSE IN THE NOTEBOOK JUST FOR SAFEKEEPING

# key_logger.py
import time
import keyboard  # pip install keyboard

log = []

print("Press 'w' for water, 'l' for lemon, 'd' for dry, 'r' for rest.")
print("Press ESC to stop.")

start_time = time.time()

while True:
    if keyboard.is_pressed("w"):
        log.append(("water", time.time() - start_time))
        print("Water logged")
        time.sleep(0.5)
    elif keyboard.is_pressed("l"):
        log.append(("lemon", time.time() - start_time))
        print("Lemon logged")
        time.sleep(0.5)
    elif keyboard.is_pressed("d"):
        log.append(("dry", time.time() - start_time))
        print("Dry swallow logged")
        time.sleep(0.5)
    elif keyboard.is_pressed("r"):
        log.append(("rest", time.time() - start_time))
        print("Rest logged")
        time.sleep(0.5)
    elif keyboard.is_pressed("esc"):
        break

# save to file
import pandas as pd
df = pd.DataFrame(log, columns=["condition", "timestamp"])
df.to_csv("annotations.csv", index=False)
```
